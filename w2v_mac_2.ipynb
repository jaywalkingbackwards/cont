```python
def remove_links(tweet):
    '''Provide a tweet and remove the links from it'''
    text_only = []
    if len(re.findall("(https://[^\s]+)", tweet)) > 0:
        tweet = re.sub("(https://[^\s]+)", "", tweet)
    if len(re.findall("(http://[^\s]+)", tweet)) > 0:
        tweet = re.sub("(http://[^\s]+)", "", tweet)    
    text_only.append(tweet)
    return text_only[0]
```

```python
import re
import nltk
nltk.download('punkt')

def text_to_sent(t):
    text = BeautifulSoup(t).text.lower()
    tokenizer = nltk.data.load('tokenizers/punkt/russian.pickle')
    raw_sentences = tokenizer.tokenize(text.strip())
    raw_sentences = [x.split(';') for x in raw_sentences]
    raw_sentences = sum(raw_sentences, [])
    sentences = [morph.lemmatize(x) for x in raw_sentences]
    return [[y for y in x if re.match('[а-яёa-z0-9]', y)]
            for x in sentences]
```

```python
from gensim.models.word2vec import Word2Vec 

num_features = 1000  # итоговая размерность вектора каждого слова
min_word_count = 5  # минимальная частотность слова, чтобы оно попало в модель
num_workers = -1     # количество ядер вашего процессора, чтоб запустить обучение в несколько потоков
context = 10        # размер окна 
downsampling = 1e-3 # внутренняя метрика модели

model = Word2Vec(sentences, workers=num_workers, size=num_features,
                 min_count=min_word_count, window=context, sample=downsampling)
```

```python
import numpy as np
%time

index2word_set = set(model.wv.index2word)

def text_to_vec(words):
    text_vec = np.zeros((1000,), dtype="float32")
    n_words = 0
    for word in words:
        if word in index2word_set:
            n_words = n_words + 1
            text_vec = np.add(text_vec, model.wv[word])
    
    if n_words != 0:
        text_vec /= n_words
    return text_vec


texts_vecs = np.zeros((len(texts), 1000), dtype="float32")

for i, text in enumerate(texts):
    s = text_to_sent(text)
    s = sum(s, [])
    texts_vecs[i] = text_to_vec(s)
```

```python
import numpy as np
%time

index2word_set2 = set(model2.wv.index2word)

def titles_to_vec(words):
    titles_vec = np.zeros((10,), dtype="float32")
    n_words = 0
    for word in words:
        if word in index2word_set2:
            n_words = n_words + 1
            titles_vec = np.add(titles_vec, model2.wv[word])
    
    if n_words != 0:
        titles_vec /= n_words
    return titles_vec

```
